## Setup

```python
# Define the URL of an image we will use later
url = "https://picsum.photos/id/237/800/600"
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown

load_dotenv()

client = OpenAI()
```

## Generate Text with the API

```python
# Set the model
MODEL = "gpt-5-nano"

# Define model inputs
INPUT = """
Your task is to write a LinkedIn post that showcases the efforts I have been making to improve my skills in AI.
Write about the latest course I am taking on RAG by Diogo Resende.
Highlight the amazing Prompt Engineering techniques that are helping me 10X my outputs.
Start with a provocative hook.
Keep paragraphs short, with only one sentence each.
"""

INSTRUCTIONS = """
You are a Founder.

Your persona is friendly and funny, with a touch of acidic humor.

Your goal is to engage and connect with your LinkedIn audience.
"""
```

```python
# Call the responses endpoint
response = client.responses.create(
    model=MODEL,
    input=INPUT,
    reasoning={
        "effort": "high"
    },
    text = {
        "verbosity": "medium"
    },
    instructions = INSTRUCTIONS
)

# Display the output
display(Markdown(response.output_text))
```

```python
# Explore the output
# response.output[0].content[0].text
response.output[1].content[0].text
```

```python
# Import the library and display the image
from IPython.display import Image
Image(url = url)
```

```python
# Describe the image with GPT
response = client.responses.create(
    model = MODEL,
    input = [
        {
            "role": "user", "content": [
                {
                    "type": "input_text", "text": "Describe the image"
                },
                {
                    "type": "input_image", "image_url": url
                }
            ]
        }
    ]
)
display(Markdown(response.output_text))
```
